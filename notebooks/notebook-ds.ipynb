```python
%config IPCompleter.use_jedi=False

import numpy as np
from gradesc import GradientDescent, MCMCMinimizer

def cost(parameters):
    target_parameters = np.array([1,1])
    return np.sum(np.square(target_parameters-parameters))

def prob(parameters):
    return np.exp(-0.5 * cost(parameters))


start_parameters = np.random.random(2)

min_parameters = [0,0]

max_parameters = [2,2]

mcmc = MCMCMinimizer(min_parameters, max_parameters)

par_hist, cost_hist = mcmc(prob, start_parameters, iterations=200)

print(mcmc.param_in_max)
```

```python
import matplotlib.pyplot as plt
plt.plot(np.log10(cost_hist))
plt.show()
```

```python
%config IPCompleter.use_jedi=False

import numpy as np
import matplotlib.pyplot as plt
from gradesc import GradientDescent, MCMC

def cost(parameters):
    target_parameters = np.array([1,1])
    return np.sum(np.square(target_parameters-parameters))

def prob(parameters):
    return np.exp(-5 * cost(parameters))


start_parameters = np.random.random(2)

min_parameters = [0,0]

max_parameters = [2,2]

mcmc = MCMC(min_parameters, max_parameters, metropolis_hastings=True, clip_limits=True)

par_hist, cost_hist = mcmc(prob, start_parameters, iterations=200, learning_rate=0.1)

print(mcmc.param_in_max, mcmc.max)
```

```python


plt.hist(par_hist[:,0], bins=100)

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(np.log10(cost_hist))
plt.show()
```
